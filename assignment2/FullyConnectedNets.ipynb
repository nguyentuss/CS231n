## Inline Question 2:

AdaGrad, like Adam, is a per-parameter optimization method that uses the following update rule:

```
cache += dw**2
w += - learning_rate * dw / (np.sqrt(cache) + eps)
```

John notices that when he was training a network with AdaGrad that the updates became very small, and that his network was learning slowly. Using your knowledge of the AdaGrad update rule, why do you think the updates would become very small? Would Adam have the same issue?


## Answer: 
Recall that with AdaGrad during the course of optimization, cache keeps the running estimate of all the squared gradients encountered during training.

During parameter update, we divide the gradient by the squared root of cache. This scaling allows the parameters with high gradients to have their learning rate reduced and the parameters with low gradients - increased.

### Explaination
1. **Reason for small updates with AdaGrad**
As the cache grows monotonically, we divide the gradient by a larger cache and produce low values. Steps towards minima get smaller which may cause updates to become very small. In convex case this is a good feature because we want steps to be small near the minima but, in non-covex case, we would stuck in a saddle point preventing further updates.
2. **Robustness of Adam** 
Adam would not have the same issue because it, similarly to RMSprop, lets the squared estimate of the gradients decay. In addition, it encorporates Momentum update rule which allows to build velocity over time and, therefore, does not cause an extreme slow-down in saddle points.
